```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("employee_turnover.csv")
data
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
x=data.drop("Employee_Turnover",axis=1)
y=data["Employee_Turnover"]
x=pd.get_dummies(x,drop_first=True)
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state=42)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
# logistic regression
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train_scaled,y_train)
y_pred=model.predict(x_test_scaled)
print("accuracy score=",accuracy_score(y_test,y_pred))
print("confusion matrix=",confusion_matrix(y_test,y_pred))
print("classification_report=",classification_report(y_test,y_pred))
```

```python
# L1 regularization
l1_lr=LogisticRegression(penalty='l1',solver='liblinear', max_iter=1000)
l1_lr.fit(x_train_scaled,y_train)
y_pred_l1=l1_lr.predict(x_test_scaled)
print("L1 accuracy=",accuracy_score(y_test,y_pred_l1))
```

```python
# L2 regularization
l2_lr=LogisticRegression(penalty='l2',solver='lbfgs',max_iter=100)
l2_lr.fit(x_train_scaled,y_train)
y_pred_l2=l2_lr.predict(x_test_scaled)
print("L2 accuracy=",accuracy_score(y_test,y_pred_l2))
```
